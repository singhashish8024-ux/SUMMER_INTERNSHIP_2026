```python
df['combined_features'] = df['Genre'] + " " + df['Lead Star'] + " " + df['Director']
df['combined_features'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert the text into numbers
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['combined_features'])

# Compute the similarity scores between all movies
similarity_scores = cosine_similarity(count_matrix)
```

```python
def get_recommendations(movie_title, top_n=5):
    # Find the row index of the requested movie
    try:
        movie_idx = df[df['Movie Name'].str.lower() == movie_title.lower()].index[0]
    except IndexError:
        return "Movie not found in the dataset. Check your spelling!"
    
    # Get all similarity scores for this movie
    sim_scores = list(enumerate(similarity_scores[movie_idx]))
    
    # Sort movies based on similarity scores (highest first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the indices of the top movies (skipping the 0th index because it's the movie itself)
    top_movies_indices = [i[0] for i in sim_scores[1:top_n+1]]
    
    # Return the titles
    return df['Movie Name'].iloc[top_movies_indices]
```

```python
get_recommendations("Fight Club")
```
