```python
# Box plot for timepoint
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['timepoint'], color='skyblue')
plt.title('Box Plot of timepoint')
plt.xlabel('timepoint')
plt.show()

# Box plot for signal
plt.figure(figsize=(8, 2))
sns.boxplot(x=df['signal'], color='salmon')
plt.title('Box Plot of signal')
plt.xlabel('signal')
plt.show()
```

```python

Q1 = df['signal'].quantile(0.25)
Q3 = df['signal'].quantile(0.75)

IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df = df[(df['signal'] >= lower_bound) & (df['signal'] <= upper_bound)]
```
