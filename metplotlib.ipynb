```python
import pandas as pd 
df = pd.read_csv('mtcars2.csv') 
import matplotlib.pyplot as plt
import numpy as np
x =["a","b","c","d","e"]
y = [10,20,30,40, 90]
z =[[1,4,2,5,3],[1,4,3,4,2]]

ax1 = plt.subplot(331)
ax2 = plt.subplot(332)
ax3 = plt.subplot(333)
ax4 = plt.subplot(334)
ax5 = plt.subplot(335)
ax6 = plt.subplot(336)
ax7 = plt.subplot(337)
ax8 = plt.subplot(338)
ax9 = plt.subplot(339)
ax1.plot(x)
ax1.set_xlabel("no of times")
ax1.set_ylabel("values")
ax1.set_title("graph 1")
ax2.stackplot(x,z)
ax2.set_xlabel("values")
ax2.set_ylabel("no of times")
ax2.set_title("graph 2")
ax3.boxplot(y)
ax3.set_xlabel("no of times")
ax3.set_ylabel("values")
ax3.set_title("graph 3")
ax4.hist(y)
ax4.set_xlabel("no of times")
ax4.set_ylabel("values")
ax4.set_title("graph 4")
ax5.bar(x,y)
ax5.set_xlabel("values")
ax5.set_ylabel("frequency")
ax5.set_title("graph 5")
ax6.violinplot(y)
ax6.set_xlabel("thickness")
ax6.set_ylabel("values")
ax6.set_title("graph 6")
ax7.barh(x,y)
ax7.set_xlabel("frequency")
ax7.set_ylabel("value")
ax7.set_title("graph 7")
ax8.scatter(x,y, s=10)
ax8.set_xlabel("values")
ax8.set_ylabel("no of times")
ax8.set_title("graph 8")
ax9.stackplot(x,y)
ax9.set_xlabel("values")
ax9.set_ylabel("no of times")
ax9.set_title("graph 9")
plt.tight_layout()
plt.show()
plt.savefig("project 4 .pdf")
```
