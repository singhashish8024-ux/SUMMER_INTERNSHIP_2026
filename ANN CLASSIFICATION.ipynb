```python
# Encode each column separately using pandas factorize
df['sex'] = pd.factorize(df['sex'])[0]
df['embarked'] = pd.factorize(df['embarked'])[0]
df['class'] = pd.factorize(df['class'])[0]
df['who'] = pd.factorize(df['who'])[0]
df['embark_town'] = pd.factorize(df['embark_town'])[0]
df['alive'] = pd.factorize(df['alive'])[0]
df['adult_male'] = pd.factorize(df['adult_male'])[0]
df['alone'] = pd.factorize(df['alone'])[0]
```

```python
df["alive"].value_counts()
```

```python

X = df.drop(columns=['survived', 'alive'])
y = df['survived']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state =123)
```

```python
 sc = StandardScaler()
 train_X = sc.fit_transform(train_X)
 test_X = sc.transform(test_X)
```

```python
model = Sequential()
model.add(Dense(units = 11, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 4, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
model.summary()
```

```python
# Change the loss function and add the metrics parameter
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print(train_X.shape)
print(train_y.shape)
```

```python
model.fit(train_X, train_y, epochs=30, batch_size=32, validation_split=0.1)

# Evaluate the model
mse = model.evaluate(test_X, test_y)
print(f'acuuracy: {mse}')
```
